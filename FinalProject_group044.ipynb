## Background and Prior Work

Recently, the outbreak of the novel coronavirus disease(COVID-19) has become a global health concern. This infectious disease was first detected in China, specifically in the city Wuhan from Hubei province. The virus then quickly spread across the whole country and later, the world. By March 1st, nearly 90000 people have been laboratory-confirmed with carrying this novel coronavirus, which leads to around three thousand deaths[1].

Understanding of COVID-19 is still limited since the problem has just emerged. By far, most information is based on previous cases about similar coronaviruses. As far as we know, It is mainly spread from person-to-person[2]. Because of the severity of the illness and yet limited knowledge of its cure, our group seeks to find the correlation between the spread of virus and factors that are associated with regional economic status. By doing so, we expect to both find an efficient governmental intervention on mitigating the spread of COVID-19, and infer possible future situations of this global concern.

In the hope of finding the best pattern to preserve citizens’ health from COVID-19, we are focusing on multiple aspects of governmental operations that can all be related back to provincial abilities/regional economic statuses to respond to sudden hazards. 

Firstly, we want to explore how economic status correlates to the virus breakout. We will measure economic status from 3 aspects: Gross regional product(abbreviated as GRP in the rest of this report) per capita, GRP growth rate from the past 3 years, and income per capita. As we learned from economic classes and from daily news digestion, GRP and residents’ income are commonly-used indicators of regional economies. Here, we use GRP and income per capita to diminish the effect of difference in population on provincial GRP results. Similar to our reasonings of choosing GRP and income per capita as indicator of economic status, GRP growth rate is another commonly-used indicator of economic strength and growth.We hypothesize that economic status is heavily connected to a city’s living conditions, including medical availability and a government that’s able to quickly adjust in response to unusual situations. A province with a more developed economy should also provide more advanced infrastructures for its residents. Therefore we hypothesize that regions with more developed economies would have a slower spread of COVID-19.  

Secondly, we want to measure population density’s correlation with virus breakout. We will measure population density by the equation density = population / area[3]. We chose to explore population density’s effect because we believe that it’s a major concern that correlates to COVID-19’s breakout rate since the virus spread largely through interpersonal contact[2]. Also, population density is an important confound that may offer alternative explanations to unexpected results from our analyses about GRP and income per capita. As we hypothesize based on our understanding of the society, in regions with higher economic status, there tends to be a higher population density as well. Due to the increased interpersonal contacts, outbursts of COVID-19 are supposedly more severe. 

Lastly, we struggled to find a way to represent the spread of COVID-19. To incorporate the increase in the number of everyday’s cumulative confirmed cases, we decided to use a function that we constructed, called max_infection_percentage to measure the breakout of virus in each province. This function calculates everyday’s cumulative number of confirmed cases in percentage of the total population in that area, and finds out the percentage in which there are the most patients confirmed in the 29 days we have records of (from February 1st to March 1st). We take that maximum percentage of infected citizens of each province as the representation of the severity of COVID-19 in that region. We will be referring to this data as “maximum infection percentage” in the rest of this report.

## Cleaning Virus Data

### Functions Used to clean Virus Data
    
The following functions were used to clean the csv files that were read in and contained various unecessary columns, characters, and such. 

```python
# remove spaces from string
def remove_space(string):
    try:
        return string.replace(" ","")
    except:
        return np.nan
    
# Find the maximum numerical number in a row
def find_max_in_a_row(df,row_index):
    num_cols = df.shape[1]
    max = 0
    for i in range(0,num_cols):
        value_in_ith_col = df.iloc[row_index,i]
        try:
            value_in_ith_col = remove_space(value_in_ith_col)
            if (int(value_in_ith_col) > max):
                max = int(value_in_ith_col)
        except:
            pass
    return max

# Find the maximum numerical numbers in each row of the dataframe, save them into df['max']
def find_max_in_a_dataframe(df):
    df['max'] = np.nan
    for i in range(0,df.shape[0]):
        df.loc[i,'max'] = find_max_in_a_row(df,i)
        
# Find the string in the data frame
# Return row_index,col_index of the string if the string is found
def find_string(df,string_to_find):
    num_rows = df.shape[0]
    num_cols = df.shape[1]
    for row_index in range(0,num_rows):
        for col_index in range(0,num_cols):
            if (df.iloc[row_index,col_index] == string_to_find):
                return row_index,col_index
    raise Exception("string_to_find not found") 
    return


# Find "Hubei" in df
# Return row_index,col_index of "Hubei" if "Hubei" is found
def find_Hubei(df):
    return find_string(df,"Hubei")

def drop_population_column(df):
    
# drop column that contains '(10,000s)'
    try:
        target_string_3 = '(10,000s)'
        target_string_3_col_index = find_string(df,target_string_3)[1]
        df = df.drop(df.columns[target_string_3_col_index],axis=1)
    except:
        pass
# drop column that contains "Population (in 10,000s)"    
    try:
        target_string_1 = "Population (in 10,000s)"
        target_string_1_col_index = find_string(df,target_string_1)[1]
        df = df.drop(df.columns[target_string_1_col_index],axis=1)
    except:
        pass
# drop column that contains "Population"
    try:
        target_string_2 = "Population"
        target_string_2_col_index = find_string(df,target_string_2)[1]
        df =df.drop(df.columns[target_string_2_col_index],axis=1)
    except:
        return df
    
# index of "Province/Region/City" = col_index of Hubei
# "Confirmed Cases" = max value in every row


def clean_df_first_pass(df):
    # find column index of "Province/Region/City"
    city_col_index = find_Hubei(df)[1]
    df["Province/Region/City"] = df.iloc[:,city_col_index]
    find_max_in_a_dataframe(df)
    # Drop all rows above Hubei
    for i in range(0, find_Hubei(df)[0]):
        df = df.drop(i)
    # Find out df["Confirmed Cases"]
    df["Confirmed Cases"] = df['max']
    df = df[["Province/Region/City","Confirmed Cases"]]
    return df



def drop_rows_whose_city_length_too_long(df,max_length):
    num_rows = df.shape[0]
    for i in range(0,num_rows):
        try:
            city = df.loc[i,'Province/Region/City']
            city_length = len(city)
            if city_length > max_length:
                df = df.drop(i)
        except:
            if not(np.isnan(city)):
                raise Exception("fail to drop the " + str(i) + "th row, which is too long")
    return df

def get_city_cases(df,region_name):
    num_rows = df.shape[0]
    return (df[df["Province/Region/City"] == region_name]["Confirmed Cases"]).iloc[0]



def merge_one_day(day):
    day_name = "Day " + str(day)
    df[day_name] = np.nan
    for i in range(0,df.shape[0]):
        #print(i)
        try:
            region_name = df.loc[i,"Province/Region/City"]
            num_cases = get_city_cases(list_of_virus_data[day],region_name)
            df.loc[i,day_name] = num_cases
        except:
            df.loc[i,day_name] = np.nan
```

```python
# Defind functions that will be used later when analysis correlation between income and maximum infection percentage

# get the virus data of one city 
def get_virus_data_of_one_city(city):
    # find row index of the city in df
    row_index = find_string(df,city)[0]
    # get data of the desired city using row index
    city_data = df.iloc[row_index,:]
    # remove city name from city_data
    city_data = city_data[1:]
    return city_data

def get_population_data_of_one_city(city):
    # find row index of the city in df
    row_index = find_string(population_from_WHO_report,city)[0]
    # get population data of the desired city using row index
    city_population = population_from_WHO_report.iloc[row_index,1]
    return city_population

def calculate_infection_percentage_of_one_city(city):
    infected_number = get_virus_data_of_one_city(city)
    population = get_population_data_of_one_city(city)
    return infected_number/population
```

## Investigate correlation between income and maximum infection percentage

### Calculate infection percentage and maximum infection percentage of each region

        infection_percentage = confirmed cases / population
        maximum infection percentage = max(infection_percentage)
These two functions above breifly explained how we obtained our maximum infection percentage. We will be using maximum infection percentage as an indicator of the spread of COVID-19 in each region in this report.

#### Funtions Caclulating infection percentage

```python
# Caclulate infection percentage
infection_percentage = df
row_index_of_regions_without_population = list()
for row in range(0,infection_percentage.shape[0]):
    city_name = infection_percentage.iloc[row,0]
    try:
        city_population = get_population_data_of_one_city(city_name)
        for column in range(1,infection_percentage.shape[1]):
            infection_percentage.iloc[row,column] = infection_percentage.iloc[row,column]/city_population
    except:
        row_index_of_regions_without_population.append(row)
for i in range (0,len(row_index_of_regions_without_population)):
    infection_percentage = infection_percentage.drop(row_index_of_regions_without_population[i])
infection_percentage = infection_percentage.reset_index(drop=True)        
```

#### Finding the maximum infection percentage


```python
# Find the maximum numerical number in a row
def find_row_max(df,row_index):
    num_cols = df.shape[1]
    max = 0.0
    for i in range(1,num_cols):
        value_in_ith_col = df.iloc[row_index,i]
        if (float(value_in_ith_col) > max):
            max = float(value_in_ith_col)
    return max

# Find the maximum numerical numbers in each row of the dataframe, save them into df['max']
def find_max(df):
    df['max'] = np.nan
    for i in range(0,df.shape[0]):
        df.loc[i,'max'] = find_row_max(df,i)

find_max(infection_percentage)
max_infection_percentage = infection_percentage[['Province/Region/City','max']]
max_infection_percentage.head()
```

We will now merge the population and area dataset by the province name. We choose to "inner" merge them because if there is a province with either no population or area, we are unable to calculate the population density. After this is completed, we can calculate the population density by dividing population by area.

*The unit for population density is number of people per square kilometer.*

```python
#merge area data and population of 2018 by region
popu_density = pd.merge(population_2018,area,on = 'Region')

#calculate the population density
popu_density['Area'] = pd.to_numeric(popu_density['Area'])
popu_density['2018'] = pd.to_numeric(popu_density['2018'])
popu_density['population density'] = popu_density['2018']/popu_density['Area']
popu_density = popu_density.drop(columns = 'Area')
popu_density = popu_density.drop(columns = '2018')
popu_density.head()
```

We will now merge the data for population density and economic status by province. 

*Clarify: Since we previously used the population data from 2018 when calculating the population density, we will also use GRP from 2018 to find the correlation between population density and economic status.

```python
#merge population density data with GRP by region
co_pd_es = pd.merge(popu_density,GRP, on = 'Region')
co_pd_es = co_pd_es[co_pd_es.columns[:3]]
#rename the 2018 GRP column
co_pd_es = co_pd_es.rename(columns = {'2018':'GRP'})
co_pd_es.head()
```

Now, we look into the correlation between population density and GRP to see if they're actually correlated. We use the Pasty statsmodel to quantify our results.

```python
co_pd_es = co_pd_es.rename(columns = {'population density':'density'})
#we want to see how well does population density and GRP correlate using pasty
outcome, predictors = patsy.dmatrices('density ~ GRP', co_pd_es)
# Now use statsmodels to intialize an OLS linear model
#  This step initializes the model, and provides the data (but not actually compute the model)
mod = sm.OLS(outcome, predictors)
# fit the model
res = mod.fit()
# Check out the results
print(res.summary())
```

As we can see, an R-squared value of 0.642 and a p-value of almost 0.000 indicates that there's a positive relationship between population density and region gross product. 

We can visualize the correlation between population density and GRP by creating a scatter plot: GRP vs. population density. After plotting, we will plot a linear regression line to show the relationship.

```python
#finding the coefficient for the fit line using polyfit
a1 = np.polyfit(co_pd_es['density'],co_pd_es['GRP'],deg = 1)[0]
b1 = np.polyfit(co_pd_es['density'],co_pd_es['GRP'],deg = 1)[1]
a1,b1
```

```python
#plot the original data
plt.scatter(co_pd_es['density'],co_pd_es['GRP'],alpha =0.3, label = 'Data')

#Generate and plot the model fit line
pred_GRP = co_pd_es['density']*a1+b1
plt.plot(co_pd_es['density'],pred_GRP,'--k',label = 'Model')

plt.title('GRP vs. Population Density by Province')
plt.xlabel("population density \n number of people per square km" )
plt.ylabel("GRP (yuan)")
plt.legend()
```

From this plot, it is clear that there is a positive relationship between population density and GRP. The regions with higher population density tend to have higher GRP. Thus, population density might also serve as a factor that confounds with the economic status. 

The provinces with higher GRP values are expected to have a slower breakout rate due to the their finer infrastructures; hospitals, clinics. However, the breakout rates tend to be higher with provinces that have a higher GRP. For this we believe that population density might be the most important factor for this claim.

We recognize that our work about COVID-19 during this special period of time may excite unnecessary problems. By comparing and contrasting the data about increase in patients from each region, we aim to find a pattern and absolutely don’t want to criticize any regions’ work in controlling the spread of coronavirus. 

Our goal is to examine and analyze potential reasons behind the quick spread of COVID-19 in relation to regional economic status. By focusing on mainland China, which is the place where coronavirus first broke out and is now slowly recovering from its impact, we aim to present a possible scenario for the rest of the world which may also suffer from this infectious disease, that some certain governmental interventions about cutting down interpersonal contacts may be useful in mitigating the spread of COVID-19.
